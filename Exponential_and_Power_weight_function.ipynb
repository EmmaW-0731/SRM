```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Exponential weight function
plt.rcParams.update({'figure.figsize':(8,5), 'figure.dpi':75})
k1=1
k2=5
k3=10
k4=15
k5=20
k6=25
#k7=100
p = np.linspace(0,1,100)
y1 = k1*(np.exp(-k1*(1-p)))/(1 - np.exp(-k1))
y2 = k2*(np.exp(-k2*(1-p)))/(1 - np.exp(-k2))
y3 = k3*(np.exp(-k3*(1-p)))/(1 - np.exp(-k3))
y4 = k4*(np.exp(-k4*(1-p)))/(1 - np.exp(-k4))
y5 = k5*(np.exp(-k5*(1-p)))/(1 - np.exp(-k5))
y6 = k6*(np.exp(-k6*(1-p)))/(1 - np.exp(-k6))
#y7 = k7*(np.exp(-k7*(1-p)))/(1 - np.exp(-k7))
fig, ax = plt.subplots()
ax.plot(p,y1,label='Absolute Risk Aversion = 1',linewidth=3)
ax.plot(p,y2,label='Absolute Risk Aversion = 5',linewidth=3)
ax.plot(p,y3,label='Absolute Risk Aversion = 10',linewidth=3)
ax.plot(p,y4,label='Absolute Risk Aversion = 15',linewidth=3)
ax.plot(p,y5,label='Absolute Risk Aversion = 20',linewidth=3)
ax.plot(p,y6,label='Absolute Risk Aversion = 25',linewidth=3)
#ax.plot(p,y7,label='Absolute Risk Aversion = 100',linewidth=3)
plt.xlabel("Cumulative probability")
plt.ylabel("Value of weighting function")

leg = ax.legend(loc='upper left', bbox_to_anchor= (0.1, -0.15), ncol=2, borderaxespad=0, frameon=False)
#plt.title('Exponential Weighting function')
```

```python
#Power weight fucntion for gamma < 1
plt.rcParams.update({'figure.figsize':(8,5), 'figure.dpi':75})
gamma1=0
gamma2=0.1
gamma3=0.3
gamma4=0.5
gamma5=0.7
gamma6=0.9
gamma7=1
p = np.linspace(0,1,100)
y1 = gamma1*(1-p)**(gamma1-1)
y2 = gamma2*(1-p)**(gamma2-1)
y3 = gamma3*(1-p)**(gamma3-1)
y4 = gamma4*(1-p)**(gamma4-1)
y5 = gamma5*(1-p)**(gamma5-1)
y6 = gamma6*(1-p)**(gamma6-1)
y7 = gamma7*(1-p)**(gamma7-1)
fig, ax = plt.subplots()
ax.plot(p,y1,label='Relative Risk Aversion = 0',linewidth=3)
ax.plot(p,y2,label='Relative Risk Aversion = 0.1',linewidth=3)
ax.plot(p,y3,label='Relative Risk Aversion = 0.3',linewidth=3)
ax.plot(p,y4,label='Relative Risk Aversion = 0.5',linewidth=3)
ax.plot(p,y5,label='Relative Risk Aversion = 0.7',linewidth=3)
ax.plot(p,y6,label='Relative Risk Aversion = 0.9',linewidth=3)
ax.plot(p,y7,label='Relative Risk Aversion = 1',linewidth=3)
plt.xlabel("Cumulative probability")
plt.ylabel("Value of risk-aversion function")

leg = ax.legend(loc='upper left', bbox_to_anchor= (0.1, -0.15), ncol=2, borderaxespad=0, frameon=False)
#plt.title('Risk Aversion Funtion: gamma < 1')
```

```python
#Power weight fucntion for gamma > 1
plt.rcParams.update({'figure.figsize':(8,5), 'figure.dpi':75})
gamma1=1.1
gamma2=1.5
gamma3=5
gamma4=20
gamma5=25
p = np.linspace(0,1,100)
y1 = gamma1*p**(gamma1-1)
y2 = gamma2*p**(gamma2-1)
y3 = gamma3*p**(gamma3-1)
y4 = gamma4*p**(gamma4-1)
y5 = gamma5*p**(gamma5-1)
fig, ax = plt.subplots()
ax.plot(p,y1,label='Relative Risk Aversion = 1.1',linewidth=3)
ax.plot(p,y2,label='Relative Risk Aversion = 1.5',linewidth=3)
ax.plot(p,y3,label='Relative Risk Aversion = 5',linewidth=3)
ax.plot(p,y4,label='Relative Risk Aversion = 20',linewidth=3)
ax.plot(p,y5,label='Relative Risk Aversion = 25',linewidth=3)
plt.xlabel("Cumulative probability")
plt.ylabel("Value of risk-aversion function")

leg = ax.legend(loc='upper left', bbox_to_anchor= (0.1, -0.15), ncol=2, borderaxespad=0, frameon=False)
#plt.title('Risk Aversion Funtion: gamma > 1')
```
