```python
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import cauchy
from scipy.stats import uniform
from scipy.stats import beta
from scipy.stats import gumbel_r
from scipy.stats import t
```

```python
#Standard normal
plt.rcParams.update({'figure.figsize':(8,5), 'figure.dpi':75})
n = 500
draws = norm.rvs(size = n)
p = np.linspace(0,1,n)
Gamma = range(0,100)
M_phi_power_norm = np.zeros(100)

for gamma in range(1,100):
   sum_uz = 0
   sum_uz_final = 0
   for i in range(len(p)):
        q = np.quantile(draws, p[i])
        u = gamma*p[i]**(gamma-1) 
        z = q
        uz= u*z
        sum_uz= sum_uz+uz
        sum_uz_final= 1/n*sum_uz
   M_phi_power_norm[gamma] = sum_uz_final
   #print(sum_uz_final)
plt.plot(Gamma,M_phi_power_norm)
```
