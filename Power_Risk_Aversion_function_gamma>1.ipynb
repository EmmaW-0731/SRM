```python
#Power Risk Aversion function: gamma>1
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({'figure.figsize':(8,5), 'figure.dpi':75})
gamma1=1.5
gamma2=5
p = np.linspace(0,1,100)
y1 = gamma1*p**(gamma1-1)
y2 = gamma2*p**(gamma2-1)
fig, ax = plt.subplots()
ax.plot(p,y1,label='Absolute Risk Aversion = 1.5')
ax.plot(p,y2,label='Absolute Risk Aversion = 5')
plt.xlabel("Cumulative probability")
plt.ylabel("Value of risk-aversion function")

leg = ax.legend(loc='upper left', bbox_to_anchor= (0.1, -0.15), ncol=2, borderaxespad=0, frameon=False)
plt.title('Risk Aversion Funtion: gamma > 1')
```
