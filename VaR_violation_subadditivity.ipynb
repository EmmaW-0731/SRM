```python
#CDF of X or Y
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.figsize':(20,10), 'figure.dpi':75})
p = np.array([ 0, 0.95, 0.96, 1])
y = np.array([ 0, 0, 100, 100])

fig, ax =plt.subplots()
plt.step(p, y)
plt.fill_between(p,y, step="pre", alpha=0.4, color='LightSkyBlue')
#ax.set_xticklabels([ 0, 0.5, 0.9, 0.95, 0.99, 0.995, 0.9998,1])
ax.set_xticks([ 0, 0.95, 0.96, 1])
plt.xlabel("CDF")
plt.ylabel("X or Y")
plt.show()
```

```python
#VaR of X or Y
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.figsize':(20,10), 'figure.dpi':75})
p = np.array([ 0, 0, 100, 100])
y = np.array([ 0, 0.96, 0.96, 1])

fig, ax =plt.subplots()
plt.step(p, y)
plt.fill_between(p,y, step="pre", alpha=0.4, color='LightSkyBlue')
#ax.set_xticklabels([ 0, 0.5, 0.9, 0.95, 0.99, 0.995, 0.9998,1])
ax.set_xticks([ 0, 100])
ax.set_yticks([ 0, 0.95, 0.96,1])
plt.xlabel("VaR(X) or VaR(Y)")
plt.ylabel("p")
plt.show()
```
