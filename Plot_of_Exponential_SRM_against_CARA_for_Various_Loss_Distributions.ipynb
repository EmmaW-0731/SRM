```python
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import cauchy
from scipy.stats import uniform
from scipy.stats import beta
from scipy.stats import gumbel_r
from scipy.stats import t
```

```python
#Standard normal
plt.rcParams.update({'figure.figsize':(8,5), 'figure.dpi':75})
n = 500
draws = norm.rvs(size = n)
p = np.linspace(0,1,n)
K = range(0,100)
M_phi_norm = np.zeros(100)

for k in range(1,100):
   sum_uz = 0
   sum_uz_final = 0
   for i in range(len(p)):
        #z = z_vals[i]  #equal to quantile 'ndtri(p)'
        #q = scipy.stats.norm.cdf(p[i], loc=0.0, scale=1.0)
        q = np.quantile(draws, p[i])
        u = np.exp(-k*(1-p[i])) 
        z = q
        uz= u*z
        sum_uz= sum_uz+uz
        #print(i, "z = %0.8f" % z, "p = %0.8f" % p, "u = %0.8f" % u,"uz = %0.8f" % uz,"sum_uz = %0.8f" % sum_uz,"sum_uz_final = %0.8f" % sum_uz_final)
   sum_uz_final= k/(1-np.exp(-k))/n*sum_uz
   M_phi_norm[k] = sum_uz_final
   #print(sum_uz_final)
plt.plot(K,M_phi_norm)
```
