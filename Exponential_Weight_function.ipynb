```python
#Exponential Weight function
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({'figure.figsize':(8,5), 'figure.dpi':75})
k1=5
k2=25
p = np.linspace(0,1,100)
y1 = k1*(np.exp(-k1*(1-p)))/(1 - np.exp(-k1))
y2 = k2*(np.exp(-k2*(1-p)))/(1 - np.exp(-k2))
fig, ax = plt.subplots()
ax.plot(p,y1,label='Absolute Risk Aversion = 5')
ax.plot(p,y2,label='Absolute Risk Aversion = 25')
plt.xlabel("Cumulative probability")
plt.ylabel("Value of weighting function")

leg = ax.legend(loc='upper left', bbox_to_anchor= (0.1, -0.15), ncol=2, borderaxespad=0, frameon=False)
plt.title('Exponential Weighting function')
```
